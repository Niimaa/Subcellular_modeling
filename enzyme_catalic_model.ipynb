```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import matplotlib
```

```python
#Declare a model which contains the system of differential equations

def enzyme_catalytic_model(s, t):
    ds = -K1f*s[0]*s[1] + K1r*s[2]
    de = -K1f*s[0]*s[1] + (K1r + K2)*s[2]
    dc = K1f*s[0]*s[1] - (K1r + K2)*s[2]
    dp = K2*s[2]
    return [ds, de, dc, dp]
```

```python
# Set constant parameter values
K1f = 0.5
K1r = 0.5
K2 = 0.5

# Specify time and initialize the floating species
time = np.linspace(0.0, 20.0, 100)
species_initial_conditions = np.array([2.0, 1.0, 2.0, 1.0])

# Solve the system of differential equations
y = odeint(enzyme_catalytic_model, species_initial_conditions, time)
```

```python
#plot the solution to system of differential equations
plt.plot(time, y[:,0], time, y[:,1], time, y[:,2], time, y[:,3], linewidth = 5)
plt.xlabel('time')
plt.ylabel('species concentrations')
plt.legend(['[s]','[e]','[c]','[p]'])
plt.show()
```
