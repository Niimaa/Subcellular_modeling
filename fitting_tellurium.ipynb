```python
import tellurium as te
import teUtils as tu

```

```python
r = te.loada('''
   J1: S1 -> S2; k1*S1
   J2: S2 -> S3; k2*S2
     
   S1 = 1; S2 = 0; S3 = 0; 
   k1 = 0.1; k2 = 0.1; # <- initial values for the fitter
''')
```

```python
r.simulate(0,20,50)
```

```python
r.plot()
```

```python
# help(teUtils.fileUtils.readCsvFile)
data, header = tu.fileUtils.readCsvFile("groundtruth_2_step_0_01.txt")
te.plotArray (data, labels=header)
```

```python
#let's fit the model
fitter = tu.model_fitter.ModelFitter(r, "groundtruth_2_step_0_01.txt", ["k1", "k2"])
```

```python
fitter.fitModel()
```

```python
print(fitter.reportFit ())

```

```python
print (fitter.minimizer.success)

```

```python
fitter.plotFitAll(figsize= (10,6))
```

```python
#for more details we can use the following code
fitter.plotFit(num_col=3, num_row=1, figsize= (18, 6))
#or we can plot all the important residuals
fitter.plotResiduals(num_col=3, num_row=1, figsize= (18,6))
```

```python
results = fitter.getFittedParameters()
print(results)
```
