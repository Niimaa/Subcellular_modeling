```python
import tellurium as te
```

```python
r = te.loada ('''
       y' = -k*y;  # Note the apostrophe
 
       y = 1; k = 0.2;
''')
m = r.simulate (0, 50, 100)
r.plot()
```

```python
r = te.loada ('''
     x' = sigma*(y - x);
     y' = x*(rho - z) - y;
     z' = x*y - beta*z;

     x = 0.96259;  y = 2.07272;  z = 18.65888;

     sigma = 10;  rho = 28; beta = 2.67;
''')

result = r.simulate (0, 20, 1000, ['time', 'x', 'y', 'z'])
r.plot (result, figsize= (11,6))

```
