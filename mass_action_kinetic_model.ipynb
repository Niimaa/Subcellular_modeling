```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from scipy.integrate import odeint

```

```python
#Declare a model which contains the system of differential equations
def mass_action_model (s, t):
    ds1 = -K1*s[0]
    ds2 = K1*s[0]-K2*s[1]
    ds3 = K2*s[1]
    return (ds1, ds2, ds3)
```

```python
# Set constant parameter values
K1 = 0.15
K2 = 0.45

# Specify time and initialize the floating species
time = np.linspace(0,20,100)
species_initial_conditions = np.array([1,0,0])

# Solve the system of differential equations
y = odeint(mass_action_model, species_initial_conditions, time)
```

```python
#plot the solution to system of differential equations
plt.plot(time, y[:,0], time, y[:,1], time, y[:,2], linewidth = 5)
plt.xlabel('time')
plt.ylabel('species concentration')
plt.legend(['[s1]','[s2]','[s3]'])
plt.show()
```
