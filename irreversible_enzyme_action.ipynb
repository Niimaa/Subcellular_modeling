```python
import tellurium as te
import teUtils
```

```python
irreversible_str = """   
      
      J1: E + S -> ES; K1*E*S;           # reaction; reaction rate low;
      J2: ES -> E + S; K_1*ES; 
      J3: ES -> E + P; K2*ES;       

      
      S = 10;                        # assign species initial conditions
      E = 2;
      ES = 0;
      Et = E + ES;
           
      
                           
     
      K1 = 2000;                     # assign constant values to global parameters
      K_1 = 100;
      K2 = 10; 
      
      

"""
```

```python
exercise_1 = te.loada(irreversible_str)
```

```python
exercise_1.simulate (0, 1, 100)

```

```python
exercise_1.plot(title= 'Irreversible enzyme action', figsize=(15,10), xtitle= 'Time', ytitle= 'Concentration', linewidth= 7)
```

```python
#rate of production
exercise_1.simulate (0, 1.4, ['time', 'J3'])
exercise_1.plot()
```

```python
velocity_str = """   
 # model mass_action                  # name the model
      compartment C1;                # specify compartments
      C1 = 1.0;                      # assign compartment volume
      species S, E, P, ES;            # specify species
      S in C1; E in C1; P in C1;     # allocate species to appropriate compartment
      ES in C1;
        
      J1: E + S -> ES; K1*E*S;
      J2: ES -> E + S; K_1*ES;
      J3: ES -> E + P; V_m * S/ (K_d + S);        # reaction; reaction rate low;

      
      S = 10;                        # assign species initial conditions
      E = 2;
      ES = 0;
      V_m = K2 * Et;
      K_d = E * S / ES;
      Et = E + ES;
      #ES = (Et * S)/(K_d + S);     
      
                           
     
      K1 = 20;                     # assign constant values to global parameters
      K_1 = 1;
      K2 = 10; 
      
      
#end
"""
```

```python
exercise_1_new = te.loada(velocity_str)
```

```python
exercise_1_new.simulate(0,2,50, ['time', 'J3'])
```
